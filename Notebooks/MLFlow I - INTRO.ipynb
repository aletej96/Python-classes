## When we install our server after installing our mlflow we now need to create a new experiment 


It is important to keep on mind that mlflow does not give us a URL only an ID in the script so we must set ours with mlflow.set_tracking_uri("") with our local port. Ex: Port 5000

```python
import mlflow
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes
from sklearn.ensemble import RandomForestRegressor

# set the experiment id

mlflow.set_tracking_uri("http://localhost:5000")
mlflow.set_experiment(experiment_id="652267140054748984")

mlflow.autolog()
```

```python


# set the experiment id
mlflow.set_experiment(experiment_id="652267140054748984")

mlflow.autolog()
db = load_diabetes()

X_train, X_test, y_train, y_test = train_test_split(db.data, db.target)

# Create and train models.
rf = RandomForestRegressor(n_estimators=100, max_depth=6, max_features=3)
rf.fit(X_train, y_train)

# Use the model to make predictions on the test dataset.
predictions = rf.predict(X_test)
```

# Summary of MLflow Execution

The code performs the following operations within an MLflow experiment:

1. **Start a Run**: Initiates a new run to record experiments.

2. **Log Tags**:
   - Within the run: Assigns a tag named `my_tag` with the value `my_value`.
   - Outside the run: Attempts to assign another tag `my_tag_2` with the value `my_value_2` (not associated with the current run).

3. **Log Parameters**:
   - `n_estimators`: Logs the number of estimators as 101.

4. **Log Metrics**:
   - `mse`: Logs the mean squared error as 5.

These operations help track and document the experiment details for future analysis and reproducibility.


```python
with mlflow.start_run():

    mlflow.set_tags({"my_tag": "my_value"})
    mlflow.log_params({"n_estimators": 101})
    mlflow.log_metrics({"mse": 5})

mlflow.set_tags({"my_tag_2": "my_value_2"})    
```
